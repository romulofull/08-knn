## **16. Selección de Características**

Método 1: Feature importance de RF

```python
# Entrenamiento de RF
model = Pipeline([
    ("scaler", StandardScaler()),
    ("knn", KNeighborsClassifier(
        n_neighbors=5,
        weights="distance",
        n_jobs=-1
    ))
])

model.fit(X_train, y_train)
```

```python
r = permutation_importance(
    model,
    X_train,
    y_train,
    n_repeats=10,
    random_state=42,
    scoring="f1_macro",  # 👈 clave para multiclase
    n_jobs=-1
)

importances = r.importances_mean * 100

df_knn_imp = (
    pd.DataFrame({
        "feature": X_train.columns,
        "knn_importance": importances
    })
    .sort_values("knn_importance", ascending=False)
)
```

```python
df_knn_imp
```

Método 2: Permutation/Shuffle importance

```python
X_train1, X_val, y_train1, y_val = train_test_split(
    X_train, y_train,
    test_size=0.2,
    random_state=42,
    stratify=y_train   # IMPORTANTE en clasificación
)

# Pipeline KNN
model_knn = Pipeline([
    ("scaler", StandardScaler()),
    ("knn", KNeighborsClassifier(
        n_neighbors=5,
        weights="distance",
        n_jobs=-1
    ))
])

# Entrenamiento
model_knn.fit(X_train1, y_train1)
```

```python
# Realizamos 10 permutaciones por cada característica
perm = permutation_importance(
    model_knn,
    X_val,
    y_val,
    n_repeats=10,
    random_state=42,
    n_jobs=-1,
    scoring="f1_macro"
)
```

```python
df_perm_imp = (
    pd.DataFrame({
        "feature": X_val.columns,
        "perm_imp": perm.importances_mean * 100
    })
    .sort_values("perm_imp", ascending=False)
)

df_perm_imp
```

Método 3: SHAP

```python
X_train1.shape
```

```python
model_lgbm = lgb.LGBMClassifier(
    objective="multiclass",
    num_class=6,
    random_state=42,
    n_jobs=-1
).fit(X_train1, y_train1)
explainer = shap.Explainer(model_lgbm, X_val)
shap_vals = explainer(
    X_val,
    check_additivity=False
)
```

```python
imp_shap = np.abs(shap_vals.values).mean(axis=(0, 2))
imp_shap_pct = imp_shap / imp_shap.sum() * 100

df_shap_imp = pd.DataFrame({
    "feature": X_val.columns,
    "shap_imp": imp_shap_pct
}).sort_values("shap_imp", ascending=False)

df_shap_imp
```

```python
shap.summary_plot(
    shap_vals.values,
    X_val,
    plot_type="bar"
)
```

Unión 3 métodos

```python
df_knn_imp = df_knn_imp.rename(columns={"knn_importance": "perm_train"})
df_perm_imp = df_perm_imp.rename(columns={"perm_imp": "perm_val"})
df_importances = (
    df_knn_imp
    .merge(df_perm_imp, on="feature", how="outer")
    .merge(df_shap_imp, on="feature", how="outer")
)
```

```python
df_importances["mean_importance"] = (
    df_importances[["perm_train", "perm_val", "shap_imp"]]
    .mean(axis=1)
)
df_importances = df_importances.sort_values(
    "mean_importance", ascending=False
)
```

## **20. Modelado**

```python
knn_no_scaling = KNeighborsClassifier(n_neighbors=3,algorithm="brute")
yhat_no_scaling = knn_no_scaling.fit(X_train[features2], y_train).predict(X_test[features2])


pipe_standard = Pipeline([
    ("scaler", StandardScaler()),
    ("knn", KNeighborsClassifier(
        n_neighbors=3,
        algorithm="brute"
    ))
])

yhat_standard = pipe_standard.fit(X_train[features2], y_train).predict(X_test[features2])

pipe_norm = Pipeline([("scaler", MinMaxScaler()), ("knn", KNeighborsClassifier(n_neighbors=3, algorithm="brute"))])
yhat_norm = pipe_norm.fit(X_train[features2], y_train).predict(X_test[features2])
```

## **21. Probar diferentes valores de K**

```python
k_values = range(1, 21)
accuracies = []

for k in k_values:
    pipe_norm = Pipeline([
        ("scaler", MinMaxScaler()),
        ("knn", KNeighborsClassifier(
            n_neighbors=k,
            algorithm="brute"
        ))
    ])


    pipe_norm.fit(X_train[features2], y_train)
    y_pred = pipe_norm.predict(X_test[features2])

    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
```

```python
best_k = k_values[accuracies.index(max(accuracies))]
best_acc = max(accuracies)

print(f"Mejor k: {best_k}")
print(f"Mejor accuracy: {best_acc:.4f}")
```

```python


plt.figure()
plt.plot(k_values, accuracies, marker="o")

plt.xlabel("Número de vecinos (k)")
plt.ylabel("Accuracy")
plt.title("Accuracy vs Número de Vecinos (k)")
plt.xticks(k_values)

plt.show()
```
